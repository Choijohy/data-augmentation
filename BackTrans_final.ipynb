```python
import csv
import pandas as pd
import ast
from tqdm import tqdm

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pyperclip as pc
from selenium.webdriver.common.keys import Keys

# from webdriver_manager.chrome import ChromeDriverManager
# driver = webdriver.Chrome(ChromeDriverManager().install())
service = Service("/usr/bin/chromedriver")



options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("disable-gpu")
options.add_argument("disable-infobars")
options.add_argument("--disable-extensions")
```

```python
def kor_to_trans(kor_data):
    # 파파고 페이지 접속
    driver.get('https://papago.naver.com/?sk=ko&tk=en')
    # 번역할 문장 삽입
    driver.find_element(By.ID, 'txtSource').send_keys(kor_data)
    # 5초간 프로세스 중지
    time.sleep(5)
    # 번역된 문장 추출
    trans = driver.find_element(By.ID, 'txtTarget').text
    # 번역창 clear
    driver.find_element(By.ID, 'txtSource').clear()
    return trans
            

        
def trans_to_kor(eng_data):
    # 파파고 페이지 접속
    driver.get('https://papago.naver.com/?sk=en&tk=ko')
    # 번역할 문장 삽입
    driver.find_element(By.ID, 'txtSource').send_keys(eng_data)
    # 5초간 프로세스 중지
    time.sleep(5)
    # 번역된 문장 추출
    backtrans =driver.find_element(By.ID, 'txtTarget').text
    # 번역창 clear
    driver.find_element(By.ID, 'txtSource').clear()
    return backtrans
```

```python
#한글 -> 영어
cnt = len(sentences)

for i in range(0,30):
    try:
        s = sentences.loc[i,'Text']
        result = kor_to_trans(s)
        sentences.loc[i,'Eng'] = result
    except:
        continue
        
for i in range(30,60):
    try:
        s = sentences.loc[i,'Text']
        result = kor_to_trans(s)
        sentences.loc[i,'Eng'] = result
    except:
        continue
    
for i in range(60,cnt):
    try:
        s = sentences.loc[i,'Text']
        result = kor_to_trans(s)
        sentences.loc[i,'Eng'] = result
    except:
        continue
```

```python
#영어 -> 한글

cnt = len(sentences)

for i in range(0,30):
    try:
        s = sentences.loc[i,'Eng']
        result = trans_to_kor(s)
        sentences.loc[i,'BackTrans'] = result
    except:
        continue
        
for i in range(30,60):
    try:
        s = sentences.loc[i,'Eng']
        result = trans_to_kor(s)
        sentences.loc[i,'BackTrans'] = result
    except:
        continue
    
for i in range(60,cnt):
    try:
        s = sentences.loc[i,'Eng']
        result = trans_to_kor(s)
        sentences.loc[i,'BackTrans'] = result
    except:
        continue
```
